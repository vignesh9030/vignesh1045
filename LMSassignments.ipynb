```python
# Declaring variables of different data types
integer_var = 42            # Integer
float_var = 3.14            # Float
string_var = "Hello, Python!"  # String
boolean_var = True          # Boolean

# Printing each variable and its type
print("Integer Variable:")
print(f"Value: {integer_var}, Type: {type(integer_var)}\n")

print("Float Variable:")
print(f"Value: {float_var}, Type: {type(float_var)}\n")

print("String Variable:")
print(f"Value: {string_var}, Type: {type(string_var)}\n")

print("Boolean Variable:")
print(f"Value: {boolean_var}, Type: {type(boolean_var)}")
```

```python
# Creating a List
my_list = [10, 20, 30, 40, 50]
print("List:", my_list)

# Accessing elements in a List
print("Accessing List elements:")
print("First element (index 0):", my_list[0])  # Index 0
print("Third element (index 2):", my_list[2])  # Index 2
print("Last element (index -1):", my_list[-1]) # Last element using -1 index
print("Slice [1:4]:", my_list[1:4])           # Slicing from index 1 to 3

print("\n")

# Creating a Tuple
my_tuple = ('apple', 'banana', 'cherry', 'date', 'elderberry')
print("Tuple:", my_tuple)

# Accessing elements in a Tuple
print("Accessing Tuple elements:")
print("Second element (index 1):", my_tuple[1])    # Index 1
print("Fourth element (index 3):", my_tuple[3])    # Index 3
print("Last element (index -1):", my_tuple[-1])    # Last element using -1 index
print("Slice [0:3]:", my_tuple[0:3])              # Slicing from index 0 to 2

print("\n")

# Creating a Dictionary
my_dict = {
    "name": "Alice",
    "age": 25,
    "city": "London",
    "job": "Engineer",
    "hobby": "Reading"
}
print("Dictionary:", my_dict)

# Accessing elements in a Dictionary
print("Accessing Dictionary elements:")
print("Value for 'name':", my_dict["name"])       # Access using key 'name'
print("Value for 'city':", my_dict["city"])       # Access using key 'city'
print("All keys:", list(my_dict.keys()))          # Get all keys
print("All values:", list(my_dict.values()))      # Get all values
```

```python
# Taking input for marks in three subjects
subject1 = float(input("Enter marks for Subject 1: "))
subject2 = float(input("Enter marks for Subject 2: "))
subject3 = float(input("Enter marks for Subject 3: "))

# Calculating the average
average = (subject1 + subject2 + subject3) / 3

# Determining the grade
if average >= 90:
    grade = "A"
elif 80 <= average < 90:
    grade = "B"
elif 70 <= average < 80:
    grade = "C"
else:
    grade = "Fail"

# Printing the result
print(f"Average Marks: {average:.2f}")
print(f"Grade: {grade}")
```

```python
# Taking input for the positive integer n
n = int(input("Enter a positive integer: "))

# Initialize the sum variable
even_sum = 0

# Loop through numbers from 1 to n
for i in range(2, n + 1, 2):  # Start from 2, increment by 2
    even_sum += i

# Print the result
print(f"The sum of all even numbers between 1 and {n} is: {even_sum}")
```

```python
# Task 1: Using a for loop and a while loop

# Ask the user to enter a positive integer
n = int(input("Enter a positive integer: "))

# 1. Using a for loop to print all numbers from 1 to n
print("Numbers from 1 to", n, ":")
for i in range(1, n + 1):
    print(i)

# 2. Using a while loop to calculate the sum of all numbers from 1 to n
sum_numbers = 0
current = 1  # Start from 1

while current <= n:
    sum_numbers += current
    current += 1

print(f"The sum of all numbers from 1 to {n} is: {sum_numbers}")
```

```python
import pandas as pd

# 1. Create a DataFrame from the provided data
data = {
    'Name': ['John', 'Alice', 'Bob', 'Diana'],
    'Age': [28, 34, 23, 29],
    'Department': ['HR', 'IT', 'Marketing', 'Finance'],
    'Salary': [45000, 60000, 35000, 50000]
}

# Create the DataFrame
df = pd.DataFrame(data)

# 2. Display the first 2 rows of the DataFrame
print("First 2 rows of the DataFrame:")
print(df.head(2))  # .head(2) returns the first 2 rows
print("\n")

# 3. Add a new column 'Bonus' where the bonus is 10% of the salary
df['Bonus'] = df['Salary'] * 0.10

# Display the updated DataFrame with the Bonus column
print("DataFrame with Bonus column:")
print(df)
print("\n")

# 4. Calculate the average salary of employees
average_salary = df['Salary'].mean()
print(f"Average salary of employees: {average_salary}\n")

# 5. Filter and display employees who are older than 25
older_than_25 = df[df['Age'] > 25]
print("Employees who are older than 25:")
print(older_than_25)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame containing the transaction data
# If the data is stored in a CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Example data structure
# Create the DataFrame with the example data
df = pd.DataFrame({
    'Transaction_Amount': [100, 200, 150, 300, 50, 400, 500, 600],
    'Account_Type': ['Savings', 'Checking', 'Savings', 'Checking', 'Savings', 'Checking', 'Savings', 'Checking'],
    'Branch': ['East', 'West', 'East', 'North', 'West', 'East', 'North', 'South']
})

# 1. Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_amount_per_account = df.groupby('Account_Type')['Transaction_Amount'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Account_Type', y='Transaction_Amount', data=total_amount_per_account, palette='Set2')
plt.title('Total Sum of Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.show()

# 2. Create a pie chart to show the percentage of transactions per Branch
branch_transaction_percentage = df.groupby('Branch')['Transaction_Amount'].sum()

plt.figure(figsize=(8, 8))
branch_transaction_percentage.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3', legend=False)
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # Hide the y-label for clarity
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the DataFrame with the example data that has the correct column names
df = pd.DataFrame({
    'Marketing_Spend': [1000, 2000, 1500, 3000, 2500],
    'Sales': [50000, 70000, 60000, 80000, 75000],
    'Drug_Effectiveness': [85, 90, 88, 92, 89],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'Age_Group': ['18-25', '26-35', '36-45', '46-60', '18-25']
})

# 1. Scatter plot for Marketing Spend vs Sales (with Region as hue)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Marketing_Spend', y='Sales', hue='Region', palette='Set2')
plt.title('Marketing Spend vs Sales by Region')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.show()

# 2. Pairplot for Marketing Spend, Sales, and Drug Effectiveness (colored by Region)
sns.pairplot(df[['Marketing_Spend', 'Sales', 'Drug_Effectiveness', 'Region']], hue='Region', palette='Set2')
plt.suptitle('Pairplot of Marketing Spend, Sales, and Drug Effectiveness', y=1.02)
plt.show()

# 3. Boxplot to compare Drug Effectiveness across Age Groups and Regions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Group', y='Drug_Effectiveness', hue='Region', palette='Set3')
plt.title('Drug Effectiveness across Age Groups and Regions')
plt.xlabel('Age Group')
plt.ylabel('Drug Effectiveness')
plt.legend(title='Region')
plt.show()

# 4. Heatmap of the correlation between Marketing Spend, Sales, and Drug Effectiveness
corr = df[['Marketing_Spend', 'Sales', 'Drug_Effectiveness']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap: Marketing Spend, Sales, and Drug Effectiveness')
plt.show()
```
